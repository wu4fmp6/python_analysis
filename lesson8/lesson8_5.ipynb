```python
import numpy as np
import pandas as pd

scores =  np.random.randint(50,101,size=(50,5))
df = pd.DataFrame(scores,
                  columns = ['國文','英文','數學','歷史','地理'],
                  index = range(1,51)
                  )
df.index.name = '學號'
df.columns.name = '科目'
df
```

```python
sum_value = df.sum(axis=1)
mean_value = df.mean(axis=1)
```

```python
df['總分'] = sum_value
df['平均'] = mean_value
```
