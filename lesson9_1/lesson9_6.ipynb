```python
import numpy as np
import pandas as pd

tips = pd.read_csv('tips.csv')
tips.columns = ['總票價','小費','吸菸者','日期','點餐','大小']
tips
```

```python
tips['小費比例'] = tips['小費']/tips['總票價']
tips
```

```python
tips.groupby('吸菸者')['小費比例'].mean()
```

```python
tips.groupby('吸菸者')['小費比例'].agg(['sum','mean'])
```

```python
df = tips.groupby('吸菸者')[['小費比例','總票價']]\
.agg([('加總','sum'),('平均','mean'),('數量','count'),('最多','max'),('最少','min'),('中間值','median'),('標準差','std')])
```

```python
df.columns.names = ['第1項','第2項']
df.index.name = '第3項'
df
```

```python
df.stack(level='第1項')
```
