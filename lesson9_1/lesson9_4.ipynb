```python
import pandas as pd

df = pd.read_csv('各鄉鎮市區人口密度.csv')
df.info()
```

```python
df1 = df.dropna()
df1.info()
```

```python
df2 = df1.rename(columns = df1.iloc[0].to_dict())
df2
```

```python
df3 = df2.drop(index=0)
df3.info()
```

```python
import numpy as np
def str_to_int(value:str) ->int:
    try:
        value_int = int(value)
    except:
        value_int = 0
    return value_int

def str_to_float(value:str) ->float:
    try:
        value_float = float(value)
    except:
        value_float = 0
    return value_float


df3['年底人口數'] = df3['年底人口數'].map(str_to_int)
df3['人口密度'] = df3['人口密度'].map(str_to_int)
df3['土地面積'] = df3['土地面積'].map(str_to_float)
```

```python
df3.info()
df3.head()
```

```python
def splite(value:str) ->str:
    if len(value) > 4 :
        return value[:3]
    else:
        return value
df3['縣市'] = df3['區域別'].map(splite)
df3
```

```python
df3.groupby('縣市')[['年底人口數','土地面積']].sum()
```

```python
df3.groupby('縣市')[['年底人口數','土地面積']].agg(['sum','mean'])
```
