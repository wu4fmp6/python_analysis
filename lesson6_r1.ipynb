```python
from bs4 import BeautifulSoup
html = '''<html>
            <body>
                <style>h1{color:blue;}</style>
                <h1>hello</h1>
            </body>
        </html>
    '''
soup = BeautifulSoup(html, 'html.parser')
style_element = soup.find_all('style')[0]
for ele in style_element:
    print(ele.string)
```
