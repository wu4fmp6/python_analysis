```python
subjects = list(student1.keys())
scores1 = list(student1.values())
scores2 = list(student2.values())

figure = plt.figure(figsize=(10, 8))
axes = figure.add_subplot(1, 1, 1)
axes.plot(subjects, scores1, 'ro-.')
axes.plot(subjects, scores2, 'bx--')
plt.show()
```

```python
subjects = list(student1.keys())
scores1 = list(student1.values())
scores2 = list(student2.values())

figure = plt.figure(figsize=(10, 8))
axes = figure.add_subplot(1, 1, 1)
axes.set_title('分數比較表')
axes.set_xlabel('科目')
axes.set_ylabel('分數')
axes.plot(subjects, scores1, 'ro-.', label='小明')
axes.plot(subjects, scores2, 'bx--', label='小華')
axes.legend()
axes.grid(axis='y')
plt.show()
```
